```python
import emoji

# Function to remove emojis from text
def remove_emojis(text):
    return emoji.replace_emoji(text, replace='')
```

```python
# Define the input and output file paths
input_file_path = os.path.join(DIRECTORY_PATH, 'test.jsonl')
output_file_path = os.path.join(DIRECTORY_PATH, 'extra_text.csv')

# Open the input and output files
with open(input_file_path, 'r', encoding='utf-8') as jsonl_file, open(output_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['text'])  # Write the header row

    # Iterate over each line in the JSONL file
    for line in jsonl_file:
        json_obj = json.loads(line.strip())
        messages = json_obj.get('messages', [])

        # Write each message to the CSV file after removing emojis
        for message in messages:
            cleaned_message = remove_emojis(message)
            csv_writer.writerow([cleaned_message])
```

```python
from os.path import exists
from melo.api import TTS

# Speed is adjustable
speed = 1.0

# You can set it manually to 'cpu' or 'cuda' or 'cuda:0' or 'mps'
device = 'auto' # Will automatically use GPU if available

# English
model = TTS(language='EN', device=device)
speaker_ids = model.hps.data.spk2id

# Read the CSV file
df = pd.read_csv(output_file_path)

# Sample Sentences Directory
sentences_path = os.mkdir(os.path.join(DIRECTORY_PATH, 'sentences'))

# Iterate over each row in the DataFrame and generate WAV files
for idx, row in df.head(10).iterrows(): #REMOVE head(10) TO DO THE ENTIRE FILE. TESTING ONLY
    text = row['text']
    unique_id = f"sentence_{idx + 1}"
    output_path = os.path.join(sentences_path, f"{unique_id}.wav")

    # Generate the WAV file
    model.tts_to_file(text, speaker_ids['EN-US'], output_path, speed=speed)
```
